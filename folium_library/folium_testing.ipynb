```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd

# Načtení CSV souboru
df = pd.read_csv('03-28-2020.csv')

# Převod sloupců na čísla a odstranění neplatných řádků
df['Lat'] = pd.to_numeric(df['Lat'], errors='coerce')
df['Long_'] = pd.to_numeric(df['Long_'], errors='coerce')
df = df.dropna(subset=['Lat', 'Long_'])  # Odstranění řádků s NaN

# Vytvoření mapy
map = folium.Map(location=[15.4529, 90.797], zoom_start=4)
folium.TileLayer('cartodbdark_matter').add_to(map)

# Vytvoření clusteru pro markery
mc = MarkerCluster()

# Procházení datového rámce a přidávání markerů
for index, row in df.iterrows():
    mc.add_child(folium.Marker(
        location=[row['Lat'], row['Long_']],
        tooltip=f"Confirmed: {row['Confirmed']}",
        icon=folium.Icon(color="red", icon="info-sign")
    ))
    folium.Circle(
        location=[row['Lat'], row['Long_']],
        color='red',
        fill=True,
        fill_color='red',
        radius=int(row['Confirmed']) * 30
    ).add_to(map)

# Přidání clusteru na mapu
map.add_child(mc)

# Uložení mapy jako HTML soubor
map.save('map_with_clusters.html')

# Zobrazen

```
